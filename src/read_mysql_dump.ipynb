```python
# https://stackoverflow.com/questions/27584405/how-to-import-a-mysqldump-into-pandas

from StringIO import StringIO
import re
import pandas as pd

def read_dump(dump_filename, target_table):
    sio = StringIO()
    count = 0
    fast_forward = True
    with open(dump_filename, 'rb') as f:
        for line in f:
            line = line.strip()
            if line.lower().startswith('insert') and target_table in line:
                fast_forward = False
            if fast_forward:
                continue
            data = re.findall('\([^\)]*\)', line)
            try:
                newline = data[0]
                ### changing here
                newlines = re.sub('\),\(', '),\n(', newline)
                ###
#                 print len(newlines)
                for newline in newlines.split('\n'):
#                     print
#                     print newline
                    newline = newline.strip(' ()')
#                     print newline
                    newline = newline.replace('`', '')
#                     print newline
                    newline = newline.replace(', ', '_ ')
#                     print newline
#                     newline = newline.replace(',N', '')
#                     new_new_line = ''
                    
#                     for item in newline:
#                         print item
                        
                    sio.write(newline)
                    sio.write("\n")
            except IndexError:
                print newline
                pass
#             if line.endswith(';'):
#                 print line
#                 break
    sio.pos = 0
    return sio
```
